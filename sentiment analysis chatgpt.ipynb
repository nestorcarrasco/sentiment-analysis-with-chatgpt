```python
import openai
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)

# Autenticación de la API de OpenAI
openai.api_key = ""

def sentiment_analysis(text):
    # Modelo de lenguaje de OpenAI que se utilizará para la clasificación de sentimientos
    model_engine = "text-davinci-003"

    # Realizar la petición a la API de OpenAI
    completions = openai.Completion.create(
        engine=model_engine,
        prompt="Clasifica el sentimiento de este texto entre Positivo, Negativo o Neutro " +text,
        max_tokens=1024,
        n=1,
        stop=None,
        temperature=0.4,
    )

    # Obtener la respuesta del modelo
    message = completions.choices[0].text.strip()
    sentiment = message.split(": ")[0]

    return sentiment

# Cargar el CSV con las frases
df = pd.read_csv('data.csv', sep=";", usecols=['Text'])

#Se aplica el modelo a la columna Text y se crea una nueva llamanda "sentimiento"
df["sentimiento"] = df["Text"].apply(sentiment_analysis)

df
```
